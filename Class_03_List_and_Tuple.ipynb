## List
Python Lists are just like dynamically sized arrays, declared in other languages. In simple language, a list is a collection of things, enclosed in [ ] and separated by commas.

The list is a sequence data type which is used to store the collection of data.

#### List Items
List items are ordered, changeable, and allow duplicate values.

```python
# List items are indexed, the first item has index [0], the second item has index [1] etc.
# Index operator []

list_1 = ['Bangladesh', 'India', 'Pakistan', 'Nepal', 18, 'July', 2022, 9.08 ]

print(list_1)
print(type(list_1))

first_content = list_1[0] # The index must be an integer.
fifth_content = list_1[4]

print('First content: ', first_content)
print('Second content: ', fifth_content)
print('\n')

print('Lenth of the list: ', len(list_1))
l = len(list_1)
print('Last content: ', list_1[l-1])
print('Last content using negative value: ', list_1[-1])
```

```python
print(list_1[-8])  # Reverse
```

### List Slicing in Python
We can access a range of items in a list by using the slicing opearator [:].

[start_number_index:(end_number_index +1)]

```python
list_1 = ['Bangladesh', 'India', 'Pakistan', 'Nepal', 18, 'July', 2022, 9.08 ]

# elements from index 2 to index 5
print(list_1[1:4])

# elements from index 4 to end
print(list_1[2:])

# elements from beginning to index 5
print(list_1[:6])

print(list_1[:])  # elements beginning to end. For vecotr and scaler.

```

```python
# List items are ordered, changeable, and allow duplicate values.
list_1[1] = 'Srilanka'

list_1[2:4] = ['America', 'Canada']

print(list_1)
print(len(list_1))
```

```python
# append() method only works for the addition of elements at the end of the List
list_1.append('abc')
print(list_1)
print(len(list_1))
```

```python
# for the addition of elements at the desired position, insert() method is used.
# Unlike append() which takes only one argument, the insert() method requires two arguments(position, value).

list_1.insert(5, "China")
print(list_1)
print(len(list_1))
```

```python
# To append elements from another list to the current list, use the extend() method.
li_1 = ['My', 'name', 'is', 'Md Emdadul Haque.']
li_2 = ['Class', 'time:', 18, 'July', 2022]

li_2.extend(li_1)
print(li_2)
```

```python
# List objects have a sort() method that will sort the list alphanumerically, ascending, by default:
li_6 = ["orange", "Mango", "kiwi", "pineapple", "banana", "3"]
li_7 = [100, 50, 65, 82, 23]

print(li_6)
li_6.sort()
print(li_6)

li_7.sort()
print(li_7)
```

```python
# # Sort the list descending

li_6.sort(reverse = True)
print(li_6)

li_7.sort(reverse = True)
print(li_7)
```

### Tuple

```python
tpl = (1,3,4,5,6)
print(tpl)
type(tpl)
```

## For Loops
- A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).
- This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.
- With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
print('My name is Md Emdadul Haque')
```

```python
result = 0
for i in range(1, 5):
    print(i)
    result = result + i

print(result)
```

```python
numbers = [3,4,6,2,54,6,3,5,7,5,3,57,5,34]
max_n = numbers[0]

for n in numbers:
    if n > max_n:
        max_n = n

print(max_n)
```

```python
li_1 = ["apple", "banana", "cherry"]
li_2 = ["mango", 'jackfruit', 'pineapple']
li_3 = ["a", "b"]

for i in li_1:
    for j in li_2:
        for k in li_3:
            print('i= ', i, '& j= ', j, '& k= ', k)
```

```python
# With the break statement we can stop the loop before it has looped through all the items
li_2 = ["mango", 'jackfruit', 'pineapple']

for i in li_2:
    #print(i)
    if i == 'jackfruit':
        break
    print(i)
```

```python
numbers = [3,4,6,2,54,6,3,5,7,5,3,57,5,34]

for i in numbers:
    #print(i)
    if i == 5:
        continue
    print(i)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        continue
    print(x)
```

## While Loop
- With the while loop we can execute a set of statements as long as a condition is true.
- The while loop requires relevant variables to be ready, in this example we need to define an indexing variable, i, which we set to 1.


```python
i = 5
while i < 5:
    print(i)
    i += 1
else:
    print("i is no longer less than 10")
```

```python
while True:
    n = int(input('Enter a number: '))
    if n == 2:
        break
    print('Square of', n, 'is', n**2)
```

```python
infinitive = ['agree','appear','arrange','ask','claim','consent',
             'decide','demand','deserve','expect','fail','forget',
             'hesitate','hope','intend','learn','manage','mean',
             'need','offer','plan','prepare','pretend','promise',
             'refuse','seem','tend','threaten','wait','want','be likely']

gerund = ['admit','appreciate','avoid','complete','consider','delay',
         'deny','discuss','enjoy','finish','keep','mention','miss',
         'postpone','practice','quit','recall','recommend','regret',
         'risk','stop','suggest','tolerate','understand']

verb_phrase = ['approve of','be better off',"can't help",'count on',
              'do not mind','forget about','get through','insist on',
              'keep on','look forward to','object to', 'think about','think on',
              'with a view to','contribute to','confess to','take to',
              'addicted to','accustomed to','get/be use to','get round to',
              'get around to','habituated to','it is no use','it is no good',
              'it is worth','it is a waste of time','would you mind']

list_f = infinitive + gerund + verb_phrase
set_f = set(list_f)

print('Write to/ing in the box below:')

true_word = []
false_word = []

for i in set_f:
    if i in infinitive:
        input_word = input('{} : '.format(i))
        index = infinitive.index(i)
        infinitive[index]='to'
        if input_word == 'to':
            print('True\n')
            true_word.append(i)
        else:
            print('False\n')
            false_word.append(i)
    elif i in gerund:
        input_word = input('{} : '.format(i))
        index = gerund.index(i)
        gerund[index]='ing'
        if input_word == 'ing':
            print('True\n')
            true_word.append(i)
        else:
            print('False\n')
            false_word.append(i)
    elif i in verb_phrase:
        input_word = input('{} : '.format(i))
        index = verb_phrase.index(i)
        verb_phrase[index]='ing'
        if input_word == 'ing':
            print('True\n')
            true_word.append(i)
        else:
            print('False\n')
            false_word.append(i)


print('\n\nTrue answer: ', len(true_word))
print('False answer: ', len(false_word))
```

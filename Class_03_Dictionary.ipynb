## Dictionary
- Dictionaries are used to store data values in key:value pairs.
- A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
- Dictionaries are written with curly brackets, and have keys and values:
- The values in dictionary items can be of any data type
- From Python's perspective, dictionaries are defined as objects with the data type 'dict':

```python
# dictionary
#a = {"A": "Apple", "B":"Ball"}
dict_1 = {'Akib':40, 'Saif Hossain': 100, 'Satil': 100, 'Antor': 60}
dict_2 = {'brand':'Ford', 'model': 'Mustang', 'year': 2022}
dict_3 = {1:"A", 2:"B", 3:"C"}

print(dict_1)
print(dict_2)
print(dict_3)

print(type(dict_1))
```

```python
print(dict_3.values())
print(dict_1.keys())
```

```python
dict_4 = {'a':(77, 88, 99), 'b':[33,53,65], 'c':[44,24,54]}
print(dict_4)
```

```python
dict_4['a']
```

```python
import pandas as pd

df = pd.DataFrame(dict_4)
print(df)
df.to_excel("A.xlsx", index=False)
```

```python
# dictionary by function
dict_5 = dict(emdad='Yes', kids='Yes', year=2026)
print(dict_5)
```

```python
# nested dictionary
training = {'CRD': {'course name': 'python', 'participants': 100},

    'Personal Course': {
        'course name': 'GIS and Remote Sensing',
        'participants': 'All'
           }}

print(training)
```

```python
training.keys()
```

```python
df1 = pd.DataFrame(training)
df1
```

```python
# join dictionary. similar to nested
CRD = {
    'course name': 'python',
    'participants': 100}

Personal_Course = {
        'course name': 'GIS and Remote Sensing',
        'participants': 'All'}

training = {'CRD': CRD,
           'Personal Course': Personal_Course}

print(training)
```

```python
# items
# Get a list of the key:value pairs

print(training.items())
```

```python
# Data acess
dict_2 = {'brand':'Ford', 'model': 'Mustang', 'year': 2022}
print(dict_2)
print(dict_2['year'])
```

```python
dict_2["model"]
```

```python
# get() it is also work for data access
print(dict_2.get('brand'))

x = dict_2.get('model')
print(x)
```

```python
# Make a change in the original dictionary, and see that the items list gets updated as well:
dict_2['model'] = 'Shah Rukh Khan'

print(dict_2)
```

```python
# Add a new item to the original dictionary, and see that the items list gets updated as well:
dict_2.update({'model':'Salman Khan'})
print(dict_2)
```

```python
dict_2["name"] = "Md Emdadul Haque"
print(dict_2)
```

```python
dict_2["name"] = "ABC"
print(dict_2)
```

```python
dict_2["X"] = "Y"
dict_2
```

```python
# The pop() method removes the item with the specified key name:
dict_2.pop('model')
print(dict_2)
```

```python
# The popitem() method removes the last inserted item
dict_2.popitem()
print(dict_2)
```

```python
# The clear() method empties the dictionary
dict_23 = dict_2.copy()
dict_23.clear()
print(dict_23)
print(type(dict_23))
```

## Python Identity Operators
Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location:

- Returns True if both variables are the same object (is) either returns False
- Returns True if both variables are not the same object (is not) either returns False

```python
name_1 = 'Emdadul'
name_2 = 'Emdadul'
name_3 = 'emdadul'

print(id(name_1))
print(id(name_2))
print(id(name_3))
```

```python
print(name_1 is name_2)
print(name_1 is name_3)
```

```python
print(name_1 is not name_2)
print(name_1 is not name_3)
```

```python
list_1 = ['Bangladesh', 'India', 'Pakistan', 'Srilanka']
list_2 = ['Bangladesh', 'India', 'Pakistan', 'Srilanka']

list_3 = ['bangladesh', 'India', 'Pakistan', 'Srilanka']

print(id(list_1[0]))   # Though list_1 and list_2 have the same content,
                    # but list_1 is not the same of list_2.
                    # Because it is a list.
print(id(list_2[0]))
print(id(list_3[0]))
```

```python
print(list_1 is list_2)
print(list_1 is list_3)
```

```python
print(id(list_1))
print(id(list_2))
```

```python
li_1 = [1, 2, 3, 4, 5]
li_2 = [1, 2, 3 , 4, 6]

print(li_1[3])
print(li_2[3])
```

```python
print(li_1 is li_2)
```

```python
print(li_1[4] is li_2[4])
```

```python
# But if we compare between single content in the both list, it returns True.
# Because it located on the same part of the memory.

print(id(list_1[0]))
print(id(list_2[0]))
```

```python
print(list_1[0] is list_2[0])
print(list_1[0] is list_3[0])
```

```python
a = 5
b = 5
c = 4
print(a is b)
print(a is c)
print('\n')
print(a is not b)
print(a is not c)
```

## Membership Operators
In Python, membership operators are used to test whether a value or variable exists within a sequence (such as a string, list, tuple, set, or dictionary). They return a Boolean value: True or False.
- in: Returns True if the specified value is present in the sequence, and False otherwise.
- not in: Returns True if the specified value is not present in the sequence, and False otherwise.

```python
list_1 = ['Bangladesh', 'India', 'Pakistan', 'Srilanka']

print('Bangladesh' in list_1)
```

```python
print('bangladesh' not in list_1)
```

```python
in_var = input("Please enter a country's name: ")

print(in_var in list_1)
```

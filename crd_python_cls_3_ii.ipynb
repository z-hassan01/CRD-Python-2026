```python
my_list = []

# Using a for loop to add numbers from 0 to 4 to the list
for i in range(5):
    my_list.append(i)

print(f"List after first loop: {my_list}")

# You can also use a loop to generate items based on a condition or calculation
another_list = []
for x in range(1, 11):
    if x % 2 == 0: # Add only even numbers
        another_list.append(x * 2) # Add double the even number

print(f"Another list: {another_list}")

# List comprehension is a more concise way to do this for simple cases
list_comprehension_example = [i * 3 for i in range(1, 6)]
print(f"List using comprehension: {list_comprehension_example}")
```

```python
years = []
for year in range(1980, 2001):  # range is exclusive of the stop value, so 2001 to include 2000
    years.append(year)

print(years)
```

```python
import random

# Define the number of years for which to generate rainfall data
num_years = 20

# Define the minimum and maximum possible annual rainfall (in mm)
min_rainfall = 100.0
max_rainfall = 2000.0

# Generate a list of random float numbers for annual rainfall
annual_rainfall = []
for _ in range(num_years):
    rainfall = random.uniform(min_rainfall, max_rainfall)
    annual_rainfall.append(round(rainfall, 2)) # Round to 2 decimal places for realism

print(f"Annual Rainfall for {num_years} years (mm): {annual_rainfall}")
```
